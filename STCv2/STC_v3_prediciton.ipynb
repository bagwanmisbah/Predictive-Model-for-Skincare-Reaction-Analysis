```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# df = pd.read_csv('your_dataset.csv')

# Define age groups
bins = [0, 29, 39, 49, 59, 69, 79, 89, 99]  # Adjust based on your data
labels = ['20s', '30s', '40s', '50s', '60s', '70s', '80s', '90s']  # Age group labels
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=True)

# Grouping by age group and ingredient, counting occurrences
ingredient_usage = df.groupby(['Age Group', 'Ingredient']).size().reset_index(name='Count')

# Pivot the table for better visualization
pivot_table = ingredient_usage.pivot(index='Age Group', columns='Ingredient', values='Count').fillna(0)

# Bar plot for ingredient usage across age groups
pivot_table.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Ingredient Usage Across Age Groups')
plt.ylabel('Count of Ingredient Usage')
plt.xlabel('Age Group')
plt.xticks(rotation=45)
plt.legend(title='Ingredient', bbox_to_anchor=(1.05, 1), loc='upper left')  # Legend outside the plot
plt.tight_layout()  # Adjust layout to make room for legend
plt.show()
```

```python
def skincare_prediction():
    # User inputs
    age = int(input("Enter your age: "))
    melanin = float(input("Enter melanin concentration (0-50): "))
    skin_type = input("Enter your skin type (Dry, Normal, Oily, Combination, Sensitive): ").capitalize()
    ingredient = input("Enter the ingredient (Kojic Acid, Vitamin C, Retinol, Hyaluronic Acid, Niacinamide, Azelaic Acid, Salicylic Acid): ").lower()
    concentration = float(input("Enter the concentration of the ingredient (%): "))

    # Initialize reaction counts
    reactions = {
        'Neutral': 0,
        'Positive': 0,
        'Brightening': 0,
        'Irritation': 0,
        'Negative': 0,
        'Recommendation': ''
    }

    # Ingredient-specific logic
    if ingredient == 'kojic acid':
        if concentration > 4:
            reactions['Irritation'] += 1
            reactions['Recommendation'] = "Exceeds safe limit; may cause irritation. Consider reducing concentration to 4% or less."
        elif concentration < 1:
            reactions['Recommendation'] = "May not observe desired effects. Recommend increasing concentration gradually up to 4%."
        else:
            if melanin > 25:
                reactions['Positive'] += 1
            else:
                reactions['Neutral'] += 1

    elif ingredient == 'vitamin c':
        if concentration > 20:
            reactions['Irritation'] += 1
            reactions['Recommendation'] = "Exceeds safe limit; may cause irritation. Consider reducing concentration to 20% or less."
        elif concentration < 8:
            reactions['Recommendation'] = "May not observe desired effects. Recommend increasing concentration gradually up to 10-20%."
        else:
            reactions['Positive'] += 1

    elif ingredient == 'retinol':
        if concentration > 1:
            reactions['Irritation'] += 1
            reactions['Recommendation'] = "Exceeds safe limit; may cause irritation. Consider reducing concentration to 1% or less."
        elif concentration < 0.2:
            reactions['Recommendation'] = "May not observe desired effects. Recommend increasing concentration gradually up to 1%."
        else:
            if age >= 30:
                reactions['Positive'] += 1

    elif ingredient == 'hyaluronic acid':
        if concentration > 2:
            reactions['Irritation'] += 1
            reactions['Recommendation'] = "Exceeds safe limit; may draw moisture from the skin. Consider reducing concentration to 2% or less."
        elif concentration < 1:
            reactions['Recommendation'] = "May not observe desired effects. Recommend increasing concentration gradually up to 1-2%."
        else:
            reactions['Positive'] += 1

    elif ingredient == 'niacinamide':
        if concentration < 5:
            reactions['Recommendation'] = "May not observe desired effects. Recommend increasing concentration gradually up to 5-10%."
        else:
            reactions['Positive'] += 1

    elif ingredient == 'azelaic acid':
        if concentration < 10:
            reactions['Recommendation'] = "May not observe desired effects. Recommend increasing concentration gradually up to 10-20%."
        elif concentration > 20:
            reactions['Irritation'] += 1
            reactions['Recommendation'] = "Exceeds safe limit; may cause irritation. Consider reducing concentration to 15-20%."
        else:
            reactions['Positive'] += 1

    elif ingredient == 'salicylic acid':
        if concentration > 2:
            reactions['Irritation'] += 1
            reactions['Recommendation'] = "Exceeds safe limit; may cause irritation. Consider reducing concentration to 2% or less."
        elif concentration < 1:
            reactions['Recommendation'] = "May not observe desired effects. Recommend increasing concentration gradually up to 2%."
        else:
            reactions['Positive'] += 1

    # Output the results
    print("\nPrediction Results:")
    for key, value in reactions.items():
        print(f"{key}: {value}")

# Call the function to test it
skincare_prediction()


# # Positive Reaction Boxplot
# plt.figure(figsize=(15, 10))
# sns.boxplot(data=df, x='Melanin Concentration', y='Positive Reaction', hue='Ingredient')
# plt.title('Positive Reactions by Melanin Concentration and Ingredients')
# plt.xlabel('Melanin Concentration')
# plt.ylabel('Positive Reaction')
# plt.show()

# # Irritation Reaction Boxplot
# plt.figure(figsize=(15, 10))
# sns.boxplot(data=df, x='Melanin Concentration', y='Irritation Reaction', hue='Ingredient')
# plt.title('Irritation Reactions by Melanin Concentration and Ingredients')
# plt.xlabel('Melanin Concentration')
# plt.ylabel('Irritation Reaction')
# plt.show()

# # Grouping reactions by ingredient
# ingredient_reaction_counts = df.groupby(['Ingredient', 'Reaction']).size().reset_index(name='Count')

# # Pivot the table for better visualization
# pivot_table = ingredient_reaction_counts.pivot(index='Ingredient', columns='Reaction', values='Count').fillna(0)

# # Bar plot for ingredient reactions
# pivot_table.plot(kind='bar', stacked=True, figsize=(12, 6))
# plt.title('Reactions by Ingredient')
# plt.ylabel('Count of Reactions')
# plt.xticks(rotation=45)
# plt.show()

```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import classification_report
import numpy as np

# Assuming your dataset is already loaded into the df variable
# and you've ensured that there are no leading or trailing spaces in column names
df.columns = df.columns.str.strip()

# Step 2: Prepare the features (X) and labels (y)
X = df[['Age', 'Melanin Concentration', 'Collagen Amount', 
        'Epidermis Thickness', 'Hydration Level', 
        'Skin Type', 'Ingredient', 'Concentration']]

# Assign the target label
y = df['Reaction']

# Step 3: Perform one-hot encoding for categorical columns (Skin Type, Ingredient)
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
X_encoded = encoder.fit_transform(X[['Skin Type', 'Ingredient']])

# Step 4: Replace the original 'Skin Type' and 'Ingredient' columns with the encoded ones
X_non_categorical = X.drop(['Skin Type', 'Ingredient'], axis=1)
X_final = np.hstack([X_non_categorical, X_encoded])

# Step 5: Split the dataset into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_final, y, test_size=0.2, random_state=42)

# Step 6: Train the RandomForestClassifier
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train)

# Step 7: Make predictions on the test set
y_pred = clf.predict(X_test)

# Step 8: Print the classification report
print(classification_report(y_test, y_pred))
```
