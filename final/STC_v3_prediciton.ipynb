### Handling Missing Values

#### By replacing with mean we fill the values for missing values attributes

```python
# Filling missing values in the 'Collagen Amount' and 'Epidermis Thickness' columns with the mean of each column
# df['Collagen Amount'].fillna(df['Collagen Amount'].mean(), inplace=True)
# df['Epidermis Thickness'].fillna(df['Epidermis Thickness'].mean(), inplace=True)

df['Collagen Amount'] = df['Collagen Amount'].fillna(df['Collagen Amount'].mean())
df['Epidermis Thickness'] = df['Epidermis Thickness'].fillna(df['Epidermis Thickness'].mean())

# Checkingg again
print("\nMissing Values after filling with mean:")
print(df.isnull().sum())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

bins = [0, 29, 39, 49, 59, 69, 79, 89, 99]  
labels = ['20s', '30s', '40s', '50s', '60s', '70s', '80s', '90s']  
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=True)

# Grouping by age group and ingredient, counting occurrences
ingredient_usage = df.groupby(['Age Group', 'Ingredient']).size().reset_index(name='Count')

pivot_table = ingredient_usage.pivot(index='Age Group', columns='Ingredient', values='Count').fillna(0)

#  ingredient usage across age groups
pivot_table.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Ingredient Usage Across Age Groups')
plt.ylabel('Count of Ingredient Usage')
plt.xlabel('Age Group')
plt.xticks(rotation=45)
plt.legend(title='Ingredient', bbox_to_anchor=(1.05, 1), loc='upper left')  
plt.tight_layout()  
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df.columns = df.columns.str.strip()

X = df[['Age', 'Melanin Concentration', 'Collagen Amount', 
        'Epidermis Thickness', 'Hydration Level', 
        'Skin Type', 'Ingredient', 'Concentration']]

y = df['Reaction']

encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
X_encoded = encoder.fit_transform(X[['Skin Type', 'Ingredient']])

X_non_categorical = X.drop(['Skin Type', 'Ingredient'], axis=1)
X_final = np.hstack([X_non_categorical, X_encoded])

X_train, X_test, y_train, y_test = train_test_split(X_final, y, test_size=0.3, random_state=42)

nb_clf = GaussianNB()
nb_clf.fit(X_train, y_train)

y_pred_nb = nb_clf.predict(X_test)

print("Naive Bayes Classification Report:")
print(classification_report(y_test, y_pred_nb))

accuracy_nb = accuracy_score(y_test, y_pred_nb)
print(f'Naive Bayes Accuracy: {accuracy_nb:.2f}')

cm_nb = confusion_matrix(y_test, y_pred_nb)
print("Naive Bayes Confusion Matrix:")
print(cm_nb)

plt.figure(figsize=(10, 7))
sns.heatmap(cm_nb, annot=True, fmt='d', cmap='Blues', xticklabels=nb_clf.classes_, yticklabels=nb_clf.classes_)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Naive Bayes Confusion Matrix')
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, StratifiedShuffleSplit
from sklearn.svm import SVC
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.exceptions import UndefinedMetricWarning
import numpy as np
import warnings


df.columns = df.columns.str.strip()

print("Class distribution:")
print(df['Reaction'].value_counts())

X = df[['Age', 'Melanin Concentration', 'Collagen Amount', 
        'Epidermis Thickness', 'Hydration Level', 
        'Skin Type', 'Ingredient', 'Concentration']]
y = df['Reaction']

encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
X_encoded = encoder.fit_transform(X[['Skin Type', 'Ingredient']])

X_non_categorical = X.drop(['Skin Type', 'Ingredient'], axis=1)
X_final = np.hstack([X_non_categorical, X_encoded])

splitter = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in splitter.split(X_final, y):
    X_train, X_test = X_final[train_index], X_final[test_index]
    y_train, y_test = y[train_index], y[test_index]

svm_model = SVC(random_state=42, class_weight='balanced')
svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)

warnings.filterwarnings("ignore", category=UndefinedMetricWarning)

print("Classification Report:")
print(classification_report(y_test, y_pred, zero_division=0))

cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, StratifiedShuffleSplit
from sklearn.svm import SVC
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.exceptions import UndefinedMetricWarning
import numpy as np
import warnings
import seaborn as sns
import matplotlib.pyplot as plt

df.columns = df.columns.str.strip()

print("Class distribution:")
print(df['Reaction'].value_counts())

X = df[['Age', 'Melanin Concentration', 'Collagen Amount', 
        'Epidermis Thickness', 'Hydration Level', 
        'Skin Type', 'Ingredient', 'Concentration']]
y = df['Reaction']

encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
X_encoded = encoder.fit_transform(X[['Skin Type', 'Ingredient']])

X_non_categorical = X.drop(['Skin Type', 'Ingredient'], axis=1)
X_final = np.hstack([X_non_categorical, X_encoded])

splitter = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in splitter.split(X_final, y):
    X_train, X_test = X_final[train_index], X_final[test_index]
    y_train, y_test = y[train_index], y[test_index]

svm_model = SVC(random_state=42, class_weight='balanced')
svm_model.fit(X_train, y_train)

y_pred = svm_model.predict(X_test)

warnings.filterwarnings("ignore", category=UndefinedMetricWarning)

print("Classification Report:")
print(classification_report(y_test, y_pred, zero_division=0))

cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False,
            xticklabels=np.unique(y), yticklabels=np.unique(y))
plt.title("Confusion Matrix")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.show()
```
